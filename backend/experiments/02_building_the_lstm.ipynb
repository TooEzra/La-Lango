## 1. Embedding Layer — turning numbers into vectors

Our tokenizer turns characters into integers (e.g. 'h' → 5).
But integers carry no meaning — is 5 closer to 4 than to 10? In language, not necessarily.

An **Embedding layer** turns each integer into a dense vector where similar tokens
end up close together in space. These vectors are learned during training.

```python
# A toy embedding: 10 possible tokens, each embedded as a 4-dimensional vector
vocab_size = 10
embed_dim  = 4

embedding = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embed_dim)

# Input: a batch of 2 sentences, each with 3 tokens
# Shape: [batch_size=2, seq_len=3]
input_indices = torch.tensor([[1, 5, 7], [2, 4, 6]])

# Output: each token becomes a vector
# Shape: [batch_size=2, seq_len=3, embed_dim=4]
embedded = embedding(input_indices)

print(f'Input shape:    {input_indices.shape}   ← [batch, seq_len]')
print(f'Embedded shape: {embedded.shape}  ← [batch, seq_len, embed_dim]')
print()
print('Token index 5 became this vector:')
print(embedded[0, 1].detach())  # batch 0, position 1 (the '5' token)
```

## 2. LSTM — processing a sequence

An LSTM (Long Short-Term Memory) reads a sequence of vectors one at a time.
At each step it updates two internal states:
- `hidden`: the short-term memory (what the model currently focuses on)
- `cell`: the long-term memory (important information from far back in the sequence)

After reading the entire sequence, the final `hidden` state is the model's
"understanding" of the whole sentence.

```python
hidden_dim = 8
lstm = nn.LSTM(input_size=embed_dim, hidden_size=hidden_dim, batch_first=True)

# Feed the embedded sequences through the LSTM
outputs, (hidden, cell) = lstm(embedded)

print(f'LSTM outputs shape: {outputs.shape}  ← [batch, seq_len, hidden_dim]')
print(f'  outputs[:,  0, :] = hidden state after reading token 0 (first character)')
print(f'  outputs[:, -1, :] = hidden state after reading the LAST token')
print()
print(f'Final hidden shape: {hidden.shape}  ← [1, batch, hidden_dim]')
print(f'Final cell shape:   {cell.shape}    ← [1, batch, hidden_dim]')
print()
print('The final hidden state is what the Encoder passes to the Decoder.')
```

## 3. The Encoder — reading the source sentence

The Encoder is just: **Embedding → LSTM → return final hidden state**.

The final hidden state is called the **context vector** — it is a compact
representation of the entire source sentence.

```python
# This is the structure the Encoder in seq2seq_lstm.py should follow
class SimpleEncoder(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)

    def forward(self, source_indices):
        # Step 1: indices → dense vectors
        embedded = self.embedding(source_indices)   # [batch, seq_len, embed_dim]
        # Step 2: read the sequence
        _, (hidden, cell) = self.lstm(embedded)     # we only keep the final state
        return hidden, cell                          # [1, batch, hidden_dim] each

encoder = SimpleEncoder(vocab_size=50, embed_dim=16, hidden_dim=32)

# Simulate a batch of 3 source sentences, each 7 characters long
source = torch.randint(1, 50, (3, 7))   # [batch=3, seq_len=7]
hidden, cell = encoder(source)

print(f'Source shape: {source.shape}  ← [batch=3, seq_len=7]')
print(f'Hidden shape: {hidden.shape}  ← [1, batch=3, hidden_dim=32]')
print()
print('The entire 7-character sentence is now compressed into a 32-dimensional vector.')
print('One vector per sentence in the batch.')
```

## 4. The Decoder — generating the translation

The Decoder takes the Encoder's hidden state and generates the translation
**one character at a time**.

At each step:
1. Takes the previous character as input (starts with SOS)
2. Updates its hidden state using the LSTM
3. Predicts the probability of each character being next

```python
class SimpleDecoder(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        # Maps the hidden state to scores for each possible next character
        self.output_layer = nn.Linear(hidden_dim, vocab_size)

    def forward_step(self, input_token, hidden, cell):
        # input_token: [batch]  →  need to add a sequence dimension
        input_token = input_token.unsqueeze(1)           # [batch, 1]
        embedded = self.embedding(input_token)            # [batch, 1, embed_dim]
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))  # [batch, 1, hidden_dim]
        prediction = self.output_layer(output.squeeze(1))  # [batch, vocab_size]
        return prediction, hidden, cell

decoder = SimpleDecoder(vocab_size=50, embed_dim=16, hidden_dim=32)

# Start decoding — SOS token index is 2
SOS_IDX = 2
batch_size = 3
sos_tokens = torch.full((batch_size,), SOS_IDX, dtype=torch.long)  # [batch]

prediction, new_hidden, new_cell = decoder.forward_step(sos_tokens, hidden, cell)

print(f'Input (SOS token) shape:  {sos_tokens.shape}   ← one token per sentence in batch')
print(f'Prediction shape:         {prediction.shape}  ← score for each of 50 possible next characters')
print()
print('The predicted next character for sentence 0:')
predicted_idx = prediction[0].argmax().item()
print(f'  argmax index = {predicted_idx}')
print(f'  (In a real model, this index maps back to an actual character via the tokenizer)')
```

## 5. The full forward pass — putting it together

During **training**, we run the decoder for each position in the target sequence.
We compare the decoder's predictions against the true target characters and
compute a loss — this tells us how wrong the model was.

**Teacher forcing** helps: instead of always feeding the decoder its own
(possibly wrong) previous prediction, we sometimes feed the correct character.
This keeps training stable early on.

```python
import random

def training_forward_pass(source, target, encoder, decoder, teacher_forcing_ratio=0.5):
    """
    This is the forward pass you will implement in Seq2SeqLSTM.forward().
    """
    batch_size   = target.shape[0]
    tgt_seq_len  = target.shape[1]
    vocab_size   = decoder.output_layer.out_features

    # Tensor to collect all predictions
    # Shape: [batch, tgt_seq_len, vocab_size]
    all_predictions = torch.zeros(batch_size, tgt_seq_len, vocab_size)

    # Step 1: Encode the source sentence
    hidden, cell = encoder(source)

    # Step 2: The first decoder input is the SOS token (target[:, 0])
    decoder_input = target[:, 0]   # SOS token for each sentence in the batch

    # Step 3: Decode one step at a time
    for t in range(1, tgt_seq_len):
        prediction, hidden, cell = decoder.forward_step(decoder_input, hidden, cell)
        all_predictions[:, t, :] = prediction

        # Teacher forcing: use the true target character OR our own prediction?
        use_teacher_forcing = random.random() < teacher_forcing_ratio
        if use_teacher_forcing:
            decoder_input = target[:, t]            # the correct character
        else:
            decoder_input = prediction.argmax(dim=1)  # our predicted character

    return all_predictions

# Simulate a small training batch
source = torch.randint(1, 50, (3, 7))   # 3 sentences, 7 source chars each
target = torch.randint(1, 50, (3, 9))   # 3 sentences, 9 target chars each
target[:, 0] = 2  # Make sure first token is SOS

all_preds = training_forward_pass(source, target, encoder, decoder)
print(f'all_predictions shape: {all_preds.shape}')
print(f'  = [batch=3, tgt_seq_len=9, vocab_size=50]')
print()
print('This tensor contains the raw scores for every position.')
print('We compare these against the true target tokens to compute the loss.')
```

## 6. Computing the loss

We use **CrossEntropyLoss** — it measures how well the model's probability
distribution matches the true label (the correct next character).

```python
loss_fn = nn.CrossEntropyLoss(ignore_index=0)  # ignore PAD tokens in the loss

# all_preds: [batch, seq_len, vocab_size]
# target:    [batch, seq_len]
#
# CrossEntropyLoss expects: [N, vocab_size] and [N]
# So we reshape by flattening the batch and sequence dimensions

# Skip position 0 (SOS) when computing loss — the model never predicts SOS
preds_flat  = all_preds[:, 1:, :].reshape(-1, 50)   # [(batch * seq_len-1), vocab_size]
targets_flat = target[:, 1:].reshape(-1)              # [(batch * seq_len-1)]

loss = loss_fn(preds_flat, targets_flat)
print(f'Loss: {loss.item():.4f}')
print()
print('This number tells us how wrong the model is on average.')
print('Lower loss = better predictions. We use backpropagation to reduce it.')
```
