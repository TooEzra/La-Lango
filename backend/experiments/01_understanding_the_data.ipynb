```python
# Add the project root to the path so we can import from lalango/
import sys
import os
# Add backend/ to path so we can import lalango
sys.path.insert(0, os.path.abspath('..'))  # go up from experiments/ to backend/

from lalango.data.cleaner import clean_sentence, clean_parallel_corpus
from lalango.data.splitter import split_corpus
from lalango.tokenizers.char_tokenizer import CharTokenizer, build_vocabulary, encode, decode

print('Imports successful!')
```

## 1. What does parallel data look like?

A parallel corpus is two files: one sentence per line, line N in source = line N in target.

```python
# A tiny toy corpus to work with in this notebook.
# In a real project, you would load this from data/raw/<lang-pair>/

source_sentences = [
    "Koso asa?",
    "Mhaka boro feel zata.",
    "Hanv school ita.",
    "To/Ti kitlo/kitli boro/boro monis asa.",
    "Aiz havaiman boro asa.",
    "Hanv tuka mog korta.",
    "Dev borem korum.",
    "Raat zali.",
]

target_sentences = [
    "How are you?",
    "I am feeling good.",
    "I am going to school.",
    "He/She is a very good person.",
    "The weather is nice today.",
    "I love you.",
    "God bless you.",
    "It is night.",
]

print(f"Total sentence pairs: {len(source_sentences)}")
print()

# Print each pair side by side
print(f"{'SOURCE':<40} TARGET")
print('-' * 70)
for src, tgt in zip(source_sentences, target_sentences):
    print(f"{src:<40} {tgt}")
```

## 2. Inspecting the data

Before cleaning, let's understand what characters and lengths we are dealing with.

```python
# Sentence length statistics
src_lengths = [len(s) for s in source_sentences]
tgt_lengths = [len(s) for s in target_sentences]

print('Source sentence lengths (in characters):')
print(f'  Min:     {min(src_lengths)}')
print(f'  Max:     {max(src_lengths)}')
print(f'  Average: {sum(src_lengths)/len(src_lengths):.1f}')
print()
print('Target sentence lengths (in characters):')
print(f'  Min:     {min(tgt_lengths)}')
print(f'  Max:     {max(tgt_lengths)}')
print(f'  Average: {sum(tgt_lengths)/len(tgt_lengths):.1f}')
```

```python
# What unique characters are in the source?
all_source_chars = set(''.join(source_sentences))
print(f'Unique characters in source ({len(all_source_chars)} total):')
print(sorted(all_source_chars))
```

## 3. Cleaning the corpus

Real-world data is messy. The cleaner handles unicode normalization,
extra whitespace, HTML tags, and filters sentences by length.

```python
# Test the cleaner on some messy strings
messy_examples = [
    "  hello   world  ",        # extra whitespace
    "<b>bold text</b>",          # HTML tags
    "café",                      # unicode (should be preserved)
]

print('Before → After cleaning:')
for text in messy_examples:
    cleaned = clean_sentence(text)
    print(f'  {repr(text):<30} → {repr(cleaned)}')
```

```python
# Clean the full corpus
clean_src, clean_tgt = clean_parallel_corpus(
    source_sentences,
    target_sentences,
    min_length=2,
    max_length=200
)

print(f'Pairs before cleaning: {len(source_sentences)}')
print(f'Pairs after cleaning:  {len(clean_src)}')
```

## 4. Splitting into train / val / test

```python
# With only 8 sentences this split is just for illustration.
# In a real project you would have thousands of pairs.
splits = split_corpus(
    clean_src,
    clean_tgt,
    train_ratio=0.7,
    val_ratio=0.15,
    test_ratio=0.15,
    seed=42
)

print('\nTraining pairs:')
for s, t in zip(splits['train']['source'], splits['train']['target']):
    print(f'  {s:<40} {t}')
```

## 5. Building the vocabulary

The tokenizer scans all training sentences, collects every unique character,
and assigns each one a number.

```python
# Build source tokenizer from training data only
# (never include val or test data in vocabulary building)
src_tokenizer = CharTokenizer()
src_tokenizer.build(splits['train']['source'])

tgt_tokenizer = CharTokenizer()
tgt_tokenizer.build(splits['train']['target'])

print(f'Source vocabulary size: {src_tokenizer.vocab_size} characters')
print(f'Target vocabulary size: {tgt_tokenizer.vocab_size} characters')
```

```python
# Look at the vocabulary mapping
print('First 20 entries in source vocabulary:')
for char, idx in sorted(src_tokenizer.char_to_idx.items(), key=lambda x: x[1])[:20]:
    print(f'  {repr(char):<12} → {idx}')
```

## 6. Encoding and decoding sentences

Now we can convert text ↔ numbers.

```python
# Pick a sentence to encode
example = 'Koso asa?'

# Encode it (no SOS for source, EOS added by default)
encoded = src_tokenizer.encode(example, add_sos=False, add_eos=True)
print(f'Original: {repr(example)}')
print(f'Encoded:  {encoded}')
print(f'Length:   {len(encoded)} (= {len(example)} chars + 1 EOS token)')
```

```python
# Decode it back
decoded = src_tokenizer.decode(encoded)
print(f'Decoded: {repr(decoded)}')
print(f'Matches original: {decoded == example}')
```

```python
# For target sentences, we add BOTH SOS and EOS
# SOS tells the decoder where to start
# EOS tells it when to stop
target_example = 'How are you?'
encoded_target = tgt_tokenizer.encode(target_example, add_sos=True, add_eos=True)

print(f'Target:  {repr(target_example)}')
print(f'Encoded: {encoded_target}')
print(f'  First token is SOS (idx 2): {encoded_target[0] == 2}')
print(f'  Last token is EOS (idx 3):  {encoded_target[-1] == 3}')
```
