# Notebook 3 — The Attention Mechanism

In the Phase 1 LSTM model, the entire source sentence is compressed into
a single hidden state vector. This creates an information bottleneck —
for long sentences, the model "forgets" words it read early on.

**Attention solves this.** Instead of relying on one fixed vector,
the decoder can look back at every encoder hidden state at each decoding step.

This notebook covers:
1. The information bottleneck problem (visualized)
2. How Bahdanau attention works, step by step
3. Visualizing attention weights — what is the model looking at?

---

## 1. The information bottleneck

In Phase 1, the encoder only passes the **final** hidden state to the decoder.
But the encoder produces a hidden state at **every position** — we are throwing most of them away.

```
Source: [H   o   w       a   r   e       y   o   u   ?]
         ↓   ↓   ↓   ↓   ↓   ↓   ↓   ↓   ↓   ↓   ↓   ↓
LSTM:  [h0, h1, h2, h3, h4, h5, h6, h7, h8, h9, h10, h11]
                                                       ↑
Phase 1: Only h11 is passed to the decoder.  ←  bottleneck!
Attention: ALL of h0..h11 are available to the decoder at every step.
```

## 2. How Bahdanau attention works

At each decoder step, we compute:

**Step 1 — Score:** How relevant is each encoder position to the current decoder state?

```
score(decoder_hidden, encoder_hidden_i) = V · tanh(W1 · decoder_hidden + W2 · encoder_hidden_i)
```

**Step 2 — Weights:** Apply softmax to turn scores into a probability distribution.

**Step 3 — Context:** Weighted sum of all encoder hidden states.

Let's implement this from scratch to see exactly what happens:

````python
# Suppose our encoder produced these hidden states (one per source character)
batch_size  = 1
src_seq_len = 5    # 5 source characters
hidden_dim  = 8

# All encoder hidden states — shape: [batch, src_seq_len, hidden_dim]
encoder_outputs = torch.randn(batch_size, src_seq_len, hidden_dim)

# The decoder's current hidden state — shape: [batch, hidden_dim]
decoder_hidden = torch.randn(batch_size, hidden_dim)

print(f'encoder_outputs shape: {encoder_outputs.shape}  ← one hidden state per source char')
print(f'decoder_hidden shape:  {decoder_hidden.shape}   ← current decoder state')
````

````python
# Step 1: Score each encoder position

W1 = nn.Linear(hidden_dim, hidden_dim)
W2 = nn.Linear(hidden_dim, hidden_dim)
V  = nn.Linear(hidden_dim, 1)

# Expand decoder_hidden to match encoder_outputs for broadcasting
# [batch, hidden_dim] → [batch, 1, hidden_dim]
decoder_hidden_expanded = decoder_hidden.unsqueeze(1)

# Compute the score for each encoder position
# Both W1(decoder) and W2(encoder) broadcast to [batch, src_seq_len, hidden_dim]
combined = torch.tanh(W1(decoder_hidden_expanded) + W2(encoder_outputs))
scores = V(combined)  # [batch, src_seq_len, 1]

print('Scores (one per source position):')
print(scores.squeeze().detach().tolist())
print(f'scores shape: {scores.shape}')
````

````python
# Step 2: Softmax → attention weights
scores_squeezed = scores.squeeze(2)   # [batch, src_seq_len]
attention_weights = torch.softmax(scores_squeezed, dim=1)

print('Attention weights (must sum to 1.0):')
print([round(w, 4) for w in attention_weights.squeeze().detach().tolist()])
print(f'Sum: {attention_weights.sum().item():.4f}')
print()
print('The decoder will pay the MOST attention to position', attention_weights.argmax().item())
````

````python
# Step 3: Context vector = weighted sum of encoder hidden states
# [batch, 1, src_seq_len] x [batch, src_seq_len, hidden_dim] → [batch, 1, hidden_dim]
context = torch.bmm(attention_weights.unsqueeze(1), encoder_outputs)
context = context.squeeze(1)   # [batch, hidden_dim]

print(f'context shape: {context.shape}  ← a single vector blending all source info')
print()
print('This context vector is used alongside the decoder hidden state')
print('to predict the next target character.')
````

## 3. The Encoder must now return ALL hidden states

In Phase 1, the Encoder returned only the final `(hidden, cell)` state.
For attention, we need the hidden state at **every position**.
This means using `outputs` from the LSTM, not just the final state.

````python
# Modified encoder that returns all hidden states
class EncoderWithAttention(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)

    def forward(self, source_indices):
        embedded = self.embedding(source_indices)
        # In Phase 1: _, (hidden, cell) = self.lstm(embedded)  ← discarded outputs
        # In Phase 2: we KEEP outputs
        outputs, (hidden, cell) = self.lstm(embedded)
        #  outputs shape: [batch, src_seq_len, hidden_dim]  ← all hidden states
        return outputs, hidden, cell

enc = EncoderWithAttention(vocab_size=50, embed_dim=16, hidden_dim=8)
source = torch.randint(1, 50, (2, 6))   # batch=2, seq_len=6

enc_outputs, enc_hidden, enc_cell = enc(source)
print(f'encoder outputs: {enc_outputs.shape}  ← [batch=2, src_seq_len=6, hidden=8]')
print(f'encoder hidden:  {enc_hidden.shape}    ← [1, batch=2, hidden=8]')
````

## 4. Visualizing attention

One of the nicest things about attention is that we can visualize it.
The attention weights show us which source words the model looked at
when generating each target word.

This is great for debugging: if the model is generating bad translations,
the attention map often shows you why.

````python
# Simulate what attention weights might look like for a good translation
# Source: "How are you?"  Target: "Koso asa?"

source_chars = list('How are you?')
target_chars = list('Koso asa?')

import random
random.seed(42)

# Simulated attention weights (in a real model, these come from the attention module)
# Each row = one target character, each column = one source character
# A high value means the decoder was looking at that source character
num_src = len(source_chars)
num_tgt = len(target_chars)

# Create a fake diagonal-ish attention pattern (roughly monotonic alignment)
import math
weights = []
for t in range(num_tgt):
    row = []
    peak = int(t * num_src / num_tgt)
    for s in range(num_src):
        # Gaussian around the expected alignment position
        w = math.exp(-0.5 * ((s - peak) ** 2))
        row.append(w)
    total = sum(row)
    weights.append([w / total for w in row])

# Print as a simple text heatmap
print('Attention heatmap (brighter = higher attention weight):')
print()
print('              ', '  '.join(source_chars))
for t_idx, row in enumerate(weights):
    bars = ''.join('█' if w > 0.25 else ('▒' if w > 0.1 else '░') for w in row)
    print(f'  target[{t_idx}] {target_chars[t_idx]}  {bars}')

print()
print('Notice how each target character mostly attends to a nearby source character.')
print('This diagonal pattern is typical of good translations.')
````

## Summary

You now understand how Bahdanau attention works:

1. **Score** every encoder position against the current decoder state
2. **Softmax** the scores to get a probability distribution (attention weights)
3. **Weighted sum** of encoder hidden states → context vector
4. Use context + decoder hidden state to predict the next character
